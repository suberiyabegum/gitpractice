```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
df=pd.read_csv(r'/content/app_user_behavior_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['rating_given'].fillna(df['rating_given'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df.head(10000)
```

```python
# Drop identifier column
df_model = df.drop(columns=['user_id'])

# Select numerical features
num_cols = df_model.select_dtypes(include=['int64', 'float64']).columns
X = df_model[num_cols]

# -------------------------------
# Data Scaling
# -------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# -------------------------------
# 2. Hierarchical Clustering
# -------------------------------
agglo = AgglomerativeClustering(n_clusters=4, linkage='ward')
df['cluster_hierarchical'] = agglo.fit_predict(X_scaled)
sil_agglo = silhouette_score(X_scaled, df['cluster_hierarchical'])
print(f"Hierarchical Silhouette Score: {sil_agglo:.3f}")
```
