```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt


# Drop identifier column
df_model = df.drop(columns=['user_id'])

# Select numerical features
num_cols = df_model.select_dtypes(include=['int64', 'float64']).columns
X = df_model[num_cols]

# -------------------------------
# Data Scaling
# -------------------------------
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import numpy as np

# 1. Scale the data
X_scaled = StandardScaler().fit_transform(X)

# 2. Define DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)

# 3. Get cluster labels
labels = dbscan.fit_predict(X_scaled)

# 4. Remove noise points (-1)
mask = labels != -1
X_filtered = X_scaled[mask]
labels_filtered = labels[mask]

# 5. Check minimum cluster requirement
if len(set(labels_filtered)) > 1:
    score = silhouette_score(X_filtered, labels_filtered)
    print("Silhouette Score:", score)
else:
    print("Silhouette score cannot be calculated (less than 2 clusters)")
```
