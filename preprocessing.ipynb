```python
import pandas as pd
import numpy as np
  
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.stats import zscore

# select numeric columns
X = df.select_dtypes(include=[np.number])

# Handle missing values
imputer = SimpleImputer(strategy="mean")
X_imputed = imputer.fit_transform(X)

# Remove outliers using z_score
z_scores =np.abs(zscore(X_imputed))
X_no_outliers = X_imputed[(z_scores < 3).all(axis=1)]

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_no_outliers)

# PCA (retain 95% variance)
pca = PCA(n_components=0.95, random_state=42)
X_pca = pca.fit_transform(X_scaled)
print("Number of features after PCA:", X_pca.shape[1])

# Find best K
best_k = 0
best_score = -1

for k in range(2,8):
    kmeans = KMeans(
        n_clusters=k,
        random_state=42,
        n_init=20,
        max_iter=500
    )
    labels = kmeans.fit_predict(X_pca)
    score = silhouette_score(X_pca, labels)
    print(f"K = {k}, Silhouette Score = {score:.4f}")

    if score > best_score:
        best_score = score
        best_k = k

print("\nBest K:", best_k)
print("Improved Silhouette Score:", round(best_score, 4))

# Final model
final_kmeans = KMeans(
    n_clusters=best_k,
    random_state=42,
    n_init=20,
    max_iter=500
)
final_labels = final_kmeans.fit_predict(X_pca)

df = df.iloc[:len(final_labels)]
df["Cluster"] = final_labels

df.head()
```

```python
cluster_counts = (
    df.groupby(['Cluster', 'cluster_name'])
      .size()
      .reset_index(name='customer_count')
)

cluster_counts
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(
    cluster_counts['cluster_name'],
    cluster_counts['customer_count']
)

plt.xlabel('Cluster Name')
plt.ylabel('Number of Customers')
plt.title('Customer Distribution Across Clusters')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Average behavioral metrics per cluster

```python
cluster_behavior = (
    df.groupby(['Cluster', 'cluster_name'])
      .agg(
          avg_sessions_per_week=('sessions_per_week', 'mean'),
          avg_session_duration=('avg_session_duration_min', 'mean'),
          avg_daily_active_minutes=('daily_active_minutes', 'mean'),
          avg_feature_clicks=('feature_clicks_per_session', 'mean'),
          avg_pages_per_session=('pages_viewed_per_session', 'mean'),
          avg_notifications_opened=('notifications_opened_per_week', 'mean'),
          avg_days_since_last_login=('days_since_last_login', 'mean'),
          avg_engagement_score=('engagement_score', 'mean'),
          avg_churn_risk=('churn_risk_score', 'mean')
      )
      .reset_index()
)

cluster_behavior
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(
    cluster_behavior['cluster_name'],
    cluster_behavior['avg_engagement_score']
)

plt.xlabel('Cluster Name')
plt.ylabel('Average Engagement Score')
plt.title('Average Engagement Score by Cluster')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Demographic & device distribution 
# Gender distribution
pd.crosstab(df['cluster_name'], df['gender'], normalize='index') * 100
```

Gender distribution is highly balanced across all user segments, with roughly equal representation of male and female users.
No cluster shows a strong gender skew, indicating that engagement differences are not driven by gender.
This suggests marketing, personalization, and retention strategies should focus on behavioral patterns rather than demographic targeting.

```python

# Device usage
pd.crosstab(df['cluster_name'], df['device_type'], normalize='index') * 100
```

Device usage is consistent across all clusters, with Android users forming the majority (~55%), followed by iOS (~35%) and Web (~10%).
No significant device-based differences are observed between engagement segments, indicating that user behavior is largely platform-agnostic.
This suggests product optimization and marketing strategies can be applied uniformly across devices, with a primary focus on mobile platforms.

```python
# subscription mix
pd.crosstab(df['cluster_name'], df['subscription_type'], normalize='index') * 100
```

Subscription distribution is remarkably similar across all engagement clusters, with Free users dominating (~60%), followed by Basic (~25%) and Premium (~15%).
High engagement is not strongly correlated with paid subscription status, indicating untapped monetization potential among active users.
These insights suggest opportunities to convert engaged Free users to paid plans through targeted value-based upsell strategies.
